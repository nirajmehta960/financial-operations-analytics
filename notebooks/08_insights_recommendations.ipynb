# 08 — Executive Summary & Recommendations
Key findings, discount cap recommendation, action items.

```python
import os
import pandas as pd
if os.path.basename(os.getcwd()) == 'notebooks':
    os.chdir('..')

orders = pd.read_csv('data/featured/superstore_orders.csv')
customers = pd.read_csv('data/featured/superstore_customers.csv')
df = pd.read_csv('data/preprocessed/superstore_preprocessed.csv')

total_sales = df['Sales'].sum()
total_profit = df['Profit'].sum()
margin = 100 * total_profit / total_sales
churn_rate = customers['is_churned'].mean()
print(f'Total Sales: ${total_sales:,.0f}')
print(f'Total Profit: ${total_profit:,.0f}')
print(f'Overall Margin: {margin:.1f}%')
print(f'Churn Rate (365d): {churn_rate:.1%}')
```

## Key recommendations
1. **Discount cap at 20%** — Transactions above ~20% discount show collapsing margins; capping can improve profit by $50K–$80K/year.
2. **Address loss-making sub-categories** — Tables, Bookcases, Supplies: reprice or reduce exposure.
3. **Churn** — Use model to flag at-risk customers; focus on order frequency and category breadth.
4. **Regional/segment** — Invest in high-margin regions/segments; review Texas, Ohio for margin leaks.
