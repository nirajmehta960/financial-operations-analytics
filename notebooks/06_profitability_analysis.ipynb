# 06 — Profitability Analysis
Sub-category P&L, discount vs profit, regional/segment comparison.

**Input:** `data/preprocessed/superstore_preprocessed.csv` and/or `data/featured/superstore_orders.csv`

```python
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
if os.path.basename(os.getcwd()) == 'notebooks':
    os.chdir('..')

df = pd.read_csv('data/preprocessed/superstore_preprocessed.csv')
subcat = df.groupby('Sub-Category').agg(Sales=('Sales', 'sum'), Profit=('Profit', 'sum')).reset_index()
subcat['margin_pct'] = subcat['Profit'] / subcat['Sales'] * 100
subcat = subcat.sort_values('Profit')
subcat
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.barh(subcat['Sub-Category'], subcat['Sales'], label='Sales', alpha=0.7)
ax.barh(subcat['Sub-Category'], subcat['Profit'], label='Profit', alpha=0.7)
ax.axvline(0, color='black', linewidth=0.5)
ax.legend()
ax.set_title('Sub-Category P&L')
plt.tight_layout()
plt.show()
```

```python
# Discount vs Profit (sample)
from src.visualizations import plot_discount_vs_profit
os.makedirs('images', exist_ok=True)
fig = plot_discount_vs_profit(df, save_path='images/discount_vs_profit.png')
plt.show()
```

```python
# Regional and segment profitability
region = df.groupby('Region').agg(Sales=('Sales', 'sum'), Profit=('Profit', 'sum')).assign(margin=lambda x: x['Profit']/x['Sales']*100)
segment = df.groupby('Segment').agg(Sales=('Sales', 'sum'), Profit=('Profit', 'sum')).assign(margin=lambda x: x['Profit']/x['Sales']*100)
print('By Region:'); print(region)
print('\nBy Segment:'); print(segment)
```
