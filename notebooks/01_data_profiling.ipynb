# 01 — Data Profiling
Initial exploration and data quality assessment for the U.S. Superstore dataset.

**Input:** `data/raw/superstore.csv`

```python
import os
import pandas as pd

if os.path.basename(os.getcwd()) == 'notebooks':
    os.chdir('..')
raw_path = 'data/raw/superstore.csv'
df = pd.read_csv(raw_path)
print(f"Shape: {df.shape}")
df.head()
```

```python
df.info()
df.describe(include='all')
```

```python
# Missing values, duplicates, date range
print('Missing:', df.isnull().sum().sum())
print('Exact duplicate rows:', df.duplicated().sum())
print('Unique Orders:', df['Order ID'].nunique(), '| Unique Customers:', df['Customer ID'].nunique())
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')
print('Date range:', df['Order Date'].min(), 'to', df['Order Date'].max())
```
